## O que vamos fazer aqui

Este é o projeto do módulo: comparar um modelo base, que só completa texto, com um modelo de
instrução, que segue pedidos, nas mesmas perguntas educacionais. A comparação usa o Ollama e
lida de forma graciosa com a ausência de qualquer um dos modelos.

## Comparando base e instrução

Ajuste os nomes dos modelos conforme o que você tem no Ollama. Para cada pergunta, enviamos
a mesma pergunta aos dois modelos e comparamos as respostas.

```python
import ollama

# Ajuste conforme os modelos que você tem. Um base completa texto; um de instrução segue pedidos.
MODELO_BASE = "llama3.1:text"        # variante base (pode não estar instalada)
MODELO_INSTRUCAO = "llama3.1"        # variante de instrução

perguntas = [
    "Explique o que é uma função matemática.",
    "Dê um exemplo de uso de derivada no dia a dia.",
]


def responder(modelo, pergunta):
    try:
        r = ollama.chat(model=modelo, messages=[{"role": "user", "content": pergunta}])
        return r["message"]["content"].strip()
    except Exception as erro:
        return f"(modelo {modelo} indisponível: {erro})"


for p in perguntas:
    print("Pergunta:", p)
    print("  Base      :", responder(MODELO_BASE, p)[:200])
    print("  Instrução :", responder(MODELO_INSTRUCAO, p)[:200])
    print()
```

Com os dois modelos disponíveis, a diferença costuma ser nítida: o de instrução vai
direto ao ponto, enquanto o base tende a continuar o texto de forma menos útil. Se você não
tem o modelo base, o de instrução ainda responde, e você pode descrever o que esperaria do
base. Para o projeto, monte uma tabela com as respostas dos dois para algumas perguntas.
