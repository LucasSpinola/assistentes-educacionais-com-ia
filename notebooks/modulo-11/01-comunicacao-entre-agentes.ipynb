## O que vamos fazer aqui

Para cooperar, agentes precisam conversar. Vamos definir um protocolo de mensagens (remetente,
tipo, conteúdo) e fazer dois agentes se comunicarem: um aluno envia uma dúvida e um tutor responde.
Python puro.

## Protocolo de mensagens e agentes

A mensagem carrega quem envia, a intenção (tipo) e o conteúdo. O tutor lê o tipo e responde no
mesmo protocolo.

```python
from dataclasses import dataclass


@dataclass
class Mensagem:
    """O protocolo de comunicação entre os agentes."""
    remetente: str
    tipo: str          # a intenção, por exemplo "duvida" ou "resposta"
    conteudo: dict


class Agente:
    def __init__(self, nome):
        self.nome = nome

    def processar(self, mensagem):
        raise NotImplementedError


class Tutor(Agente):
    def processar(self, mensagem):
        if mensagem.tipo == "duvida":
            tema = mensagem.conteudo.get("tema", "")
            explicacao = f"Vamos lá. {tema} é um conceito importante, deixa eu explicar."
            return Mensagem(self.nome, "resposta", {"explicacao": explicacao})
        return Mensagem(self.nome, "erro", {"motivo": "não sei tratar este tipo"})


tutor = Tutor("tutor")
pergunta = Mensagem("aluno", "duvida", {"tema": "a derivada"})
resposta = tutor.processar(pergunta)

print("Mensagem do aluno:", pergunta)
print("Resposta do tutor:", resposta)
```

O aluno envia uma dúvida e o tutor responde, ambos seguindo o mesmo protocolo (sabem ler
remetente, tipo e conteúdo). Esse contrato comum é o que permite coordenar vários agentes sem que
um conheça os detalhes do outro. Para o projeto, faça dois agentes manterem uma troca coerente.
