## O que vamos fazer aqui

Vamos fechar a conta que ficou aberta no Módulo 1, onde o perceptron falhou no XOR.
Aqui construímos uma rede neural com uma camada escondida e a treinamos com
backpropagation feita à mão, vendo-a resolver o XOR que o perceptron não conseguia.

É a engrenagem básica de todo o Deep Learning, multiplicações de matrizes, uma
ativação não linear e o gradiente. Só numpy.

## A rede e o treino com backpropagation

A rede tem uma camada escondida com ativação tanh e uma saída com sigmoide. O laço faz
o passo para frente (prever) e o passo para trás (propagar o erro e atualizar os
pesos), repetidos muitas vezes.

```python
import numpy as np


def sigmoide(z):
    return 1 / (1 + np.exp(-z))


# As quatro combinações do XOR.
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=float)
y = np.array([[0], [1], [1], [0]], dtype=float)

rng = np.random.default_rng(0)
H = 8                                   # neurônios na camada escondida
W1 = rng.normal(0, 1, (2, H)); b1 = np.zeros((1, H))
W2 = rng.normal(0, 1, (H, 1)); b2 = np.zeros((1, 1))
taxa = 0.5

for epoca in range(20000):
    # Passo para frente.
    a1 = np.tanh(X @ W1 + b1)           # camada escondida com ativação não linear
    a2 = sigmoide(a1 @ W2 + b2)         # saída como probabilidade

    # Passo para trás (backpropagation).
    d2 = a2 - y                          # erro na saída
    d1 = (d2 @ W2.T) * (1 - a1**2)       # erro propagado, com a derivada do tanh

    # Atualiza todos os pesos por gradiente descendente.
    W2 -= taxa * a1.T @ d2 / len(X); b2 -= taxa * d2.sum(0, keepdims=True) / len(X)
    W1 -= taxa * X.T @ d1 / len(X); b1 -= taxa * d1.sum(0, keepdims=True) / len(X)

saida = sigmoide(np.tanh(X @ W1 + b1) @ W2 + b2)
print("Previsto:", (saida > 0.5).astype(int).ravel())
print("Esperado:", y.astype(int).ravel())
```

A rede prevê `[0 1 1 0]`, acertando as quatro combinações do XOR. A camada
escondida criou características intermediárias que tornam o problema linearmente
separável na saída, exatamente o que faltava ao perceptron. Para o projeto, aplique a
mesma rede a outro problema não linear, como dois círculos concêntricos.
