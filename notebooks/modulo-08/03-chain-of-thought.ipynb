## O que vamos fazer aqui

O Chain of Thought pede ao modelo que raciocine passo a passo antes de responder, o que melhora
muito os problemas de várias etapas. Vamos montar um prompt direto e um prompt com raciocínio
para o mesmo problema e compará-los via Ollama. A montagem é Python puro; o envio é opcional.

## Montando os dois prompts

O prompt direto cobra logo a resposta. O Chain of Thought zero-shot acrescenta a frase mágica
"vamos pensar passo a passo", que faz o modelo desenvolver o raciocínio.

```python
def prompt_direto(problema):
    return f"{problema}\nResponda apenas com o número final."


def prompt_cot(problema):
    return f"{problema}\nVamos pensar passo a passo e só então dar a resposta final."


problema = (
    "Uma turma tem 28 alunos. Cada aluno precisa de 3 cadernos, e cada pacote traz "
    "4 cadernos. Quantos pacotes a turma precisa comprar?"
)

print("PROMPT DIRETO:\n", prompt_direto(problema))
print("\nPROMPT CHAIN OF THOUGHT:\n", prompt_cot(problema))
```

Os dois pedem a mesma coisa. O direto exige acertar de uma vez (arriscado neste
problema de três etapas). O Chain of Thought pede o desenvolvimento, e o modelo costuma chegar
a 28x3=84 cadernos, depois 84/4=21, logo 21 pacotes.

## Caminho opcional, comparar as respostas

```python
try:
    import ollama

    for nome, p in [("DIRETO", prompt_direto(problema)), ("CHAIN OF THOUGHT", prompt_cot(problema))]:
        r = ollama.chat(model="llama3.1", messages=[{"role": "user", "content": p}])
        print(f"[{nome}]\n{r['message']['content'].strip()}\n")
except Exception as erro:
    print("Ollama não disponível. Veja docs/setup.md. Detalhe:", erro)
```

A versão com raciocínio costuma errar bem menos nesse tipo de problema. Para o projeto,
compare a taxa de acerto das duas formas em vários problemas de múltiplas etapas.
