## O que vamos fazer aqui

No zero-shot, pedimos uma tarefa sem dar exemplos. A qualidade depende muito de como pedimos.
Vamos montar dois prompts para a mesma tarefa, um vago e um bem estruturado, e comparar as
respostas de um LLM via Ollama. A montagem dos prompts é Python puro; o envio ao modelo é
opcional.

## Montando os prompts

Um bom prompt zero-shot deixa claros o papel, a tarefa, o contexto e o formato. Compare a
diferença entre pedir de forma solta e de forma estruturada.

```python
def prompt_vago(conceito):
    """Um pedido solto, sem papel, nível ou formato."""
    return f"Explique {conceito}."


def prompt_estruturado(conceito, nivel, formato):
    """Um prompt zero-shot com papel, tarefa, contexto e formato bem definidos."""
    return (
        "Você é um tutor paciente e didático.\n"
        f"Tarefa: explique o conceito de {conceito}.\n"
        f"Contexto: o aluno está no nível {nivel}.\n"
        f"Formato: {formato}.\n"
        "Use linguagem simples e um exemplo do dia a dia."
    )


print("PROMPT VAGO:\n", prompt_vago("derivada"))
print("\nPROMPT ESTRUTURADO:\n",
      prompt_estruturado("derivada", "iniciante", "no máximo três frases"))
```

O prompt vago dá total liberdade ao modelo, e a resposta tende a ser genérica. O
estruturado guia o modelo com papel, nível e formato, e a resposta tende a ser muito mais
útil.

## Caminho opcional, enviar ao LLM

Esta célula envia os dois prompts ao Ollama para comparar as respostas de verdade. Roda apenas
se o Ollama estiver disponível.

```python
try:
    import ollama

    for nome, prompt in [("VAGO", prompt_vago("derivada")),
                         ("ESTRUTURADO", prompt_estruturado("derivada", "iniciante", "três frases"))]:
        r = ollama.chat(model="llama3.1", messages=[{"role": "user", "content": prompt}])
        print(f"[{nome}] {r['message']['content'].strip()}\n")
except Exception as erro:
    print("Ollama não disponível. Veja docs/setup.md. Detalhe:", erro)
```

A resposta ao prompt estruturado costuma ser bem mais adequada ao aluno. Para o
projeto, monte uma pequena biblioteca de prompts zero-shot para tarefas de um tutor.
