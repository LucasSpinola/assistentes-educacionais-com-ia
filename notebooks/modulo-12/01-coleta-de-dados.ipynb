## O que vamos fazer aqui

Tudo em Learning Analytics começa pela coleta estruturada de eventos. Vamos definir um esquema de
evento (aluno, instante, tipo, detalhes), montar uma pequena coleção e consultá-la por aluno e por
tipo. Python puro.

## Estruturando e consultando eventos

Um esquema fixo é o que torna os dados utilizáveis. Cada evento tem sempre os mesmos campos, o que
permite agregá-los depois.

```python
from dataclasses import dataclass
from collections import Counter


@dataclass
class Evento:
    """Um evento de aprendizagem, com esquema fixo."""
    aluno: str
    instante: int
    tipo: str
    detalhes: dict


eventos = [
    Evento("ana", 1, "duvida", {"tema": "derivada"}),
    Evento("ana", 2, "resposta", {"tema": "derivada", "correto": True}),
    Evento("ana", 3, "resposta", {"tema": "derivada", "correto": False}),
    Evento("bruno", 1, "acesso_material", {"tema": "matriz"}),
    Evento("bruno", 2, "resposta", {"tema": "matriz", "correto": True}),
]


def eventos_do_aluno(eventos, aluno):
    return [e for e in eventos if e.aluno == aluno]


def resumo_tipos(eventos, aluno):
    return Counter(e.tipo for e in eventos_do_aluno(eventos, aluno))


print("Eventos da Ana:", len(eventos_do_aluno(eventos, "ana")))
print("Tipos de evento da Ana:", dict(resumo_tipos(eventos, "ana")))
print("Tipos de evento do Bruno:", dict(resumo_tipos(eventos, "bruno")))
```

A coleção estruturada permite consultar facilmente os eventos de cada aluno e resumir os
tipos de ação. Esse esquema simples sustenta todas as métricas e modelos do módulo. Para o projeto,
monte um coletor e reflita sobre quais dados coletar de forma ética.
