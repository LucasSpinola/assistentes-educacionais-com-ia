## O que vamos fazer aqui

Antes de qualquer modelo de linguagem, é preciso quebrar o texto em pedaços, os
tokens. Vamos tokenizar perguntas de alunos de duas formas e comparar: a separação
ingênua por espaços, que tropeça na pontuação e nas maiúsculas, e um tokenizador
melhor, que normaliza o texto e separa a pontuação.

Esse mesmo conjunto de perguntas vai reaparecer nas próximas aulas, ganhando uma
etapa de processamento de cada vez. O notebook é Python puro, com a biblioteca `re`.

## Dois tokenizadores

O tokenizador ingênuo só faz `split()` pelos espaços. O melhor passa o texto para
minúsculas e usa uma expressão regular que separa sequências de letras e números da
pontuação. Compare a saída dos dois na primeira pergunta.

```python
import re

perguntas = [
    "Como faço a derivada de uma função?",
    "Qual é a regra da cadeia na derivada?",
    "Como resolvo um sistema linear com matrizes?",
    "O que é um autovetor de uma matriz?",
    "Como declaro uma função em Python?",
    "O que é um laço de repetição em Python?",
]


def tokenizar_ingenuo(texto):
    """Separa apenas pelos espaços, sem tratar pontuação ou maiúsculas."""
    return texto.split()


def tokenizar(texto):
    """Normaliza para minúsculas e separa palavras e pontuação."""
    return re.findall(r"\w+|[^\w\s]", texto.lower(), re.UNICODE)


print("Ingênuo:", tokenizar_ingenuo(perguntas[0]))
print("Melhor :", tokenizar(perguntas[0]))
```

No tokenizador ingênuo, a interrogação fica grudada em `função?`, e uma
palavra com maiúscula contaria separado da sua versão minúscula. O tokenizador melhor
separa a pontuação como token próprio e unifica a caixa, deixando tudo mais
consistente.

## Efeito sobre o vocabulário

A diferença entre os dois aparece com clareza no tamanho do vocabulário, ou seja, no
número de tokens distintos. A pontuação grudada e as maiúsculas inflam o vocabulário
do tokenizador ingênuo.

```python
vocab_ingenuo = set()
vocab_bom = set()
for p in perguntas:
    vocab_ingenuo.update(tokenizar_ingenuo(p))
    vocab_bom.update(tokenizar(p))

print("Tokens distintos, ingênuo:", len(vocab_ingenuo))
print("Tokens distintos, melhor :", len(vocab_bom))
print("\nExemplos com pontuação grudada (ingênuo):")
print([t for t in vocab_ingenuo if not t.isalnum()][:6])
```

O vocabulário do tokenizador melhor tende a ser mais enxuto e limpo, porque
junta formas que o ingênuo separava à toa. Para o projeto da aula, torne o
tokenizador configurável (manter ou não a pontuação, converter ou não para
minúsculas) e mostre, em uma tabela, como o tamanho do vocabulário muda com cada
opção.
