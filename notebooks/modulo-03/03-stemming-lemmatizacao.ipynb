## O que vamos fazer aqui

A mesma ideia aparece em muitas formas: derivar, derivada, derivadas, derivando. Para
que essas variações contem como uma coisa só, vamos reduzir cada palavra a um radical
comum com um stemmer simples de regras para o português. O objetivo não é um stemmer
perfeito, e sim entender o mecanismo e ver as variações convergirem. Python puro.

## Um stemmer simples

A estratégia é cortar terminações conhecidas, testando primeiro os sufixos mais
longos. A condição de tamanho mínimo evita cortar palavras curtas demais e virar
radicais sem sentido.

```python
palavras = [
    "derivada", "derivadas", "derivar", "derivando",
    "matriz", "matrizes",
    "função", "funções",
    "linear", "lineares",
    "repetição", "repetições",
]

SUFIXOS = ["ações", "ação", "antes", "ando", "ares", "adas", "ada", "es", "as", "os", "s"]


def stemmer_simples(palavra):
    """Remove um sufixo conhecido para aproximar o radical da palavra."""
    for sufixo in SUFIXOS:
        if palavra.endswith(sufixo) and len(palavra) - len(sufixo) >= 3:
            return palavra[: -len(sufixo)]
    return palavra


for p in palavras:
    print(f"{p:12} -> {stemmer_simples(p)}")
```

Veja palavras como derivada, derivadas e derivar se aproximarem de um radical
comum, e matriz e matrizes se juntarem pelo plural. Observe também algum exagero, pois
um stemmer de regras erra de vez em quando, o que motiva métodos mais elaborados como
o RSLP e a lemmatização com spaCy.

## Agrupando variações pelo radical

Para enxergar o efeito, agrupamos as palavras pelo radical que o stemmer produz. Cada
grupo reúne as formas que passariam a contar juntas.

```python
from collections import defaultdict

grupos = defaultdict(list)
for p in palavras:
    grupos[stemmer_simples(p)].append(p)

for radical, formas in grupos.items():
    print(f"{radical:10} <- {formas}")
```

Cada radical reúne várias formas, o que reduz o tamanho do vocabulário e
concentra as contagens. Para o projeto, monte o vocabulário das perguntas sem
normalização e com o stemmer, compare os tamanhos e discuta o ganho de unir variações
e o risco de unir palavras que não deveriam se juntar.
