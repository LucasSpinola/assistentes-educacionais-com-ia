## O que vamos fazer aqui

Nem toda palavra tem o mesmo valor. Artigos e preposições, as stopwords, aparecem em
toda parte e dizem pouco sobre o assunto. Vamos contar a frequência das palavras nas
perguntas dos alunos, ver a lei de Zipf em ação, e depois remover as stopwords para
deixar à mostra os termos de conteúdo. Tudo em Python puro.

## Frequência das palavras

Primeiro contamos quantas vezes cada palavra aparece, ignorando a pontuação. A lei de
Zipf prevê que poucas palavras vão concentrar a maior parte das ocorrências, e elas
costumam ser justamente as stopwords.

```python
import re
from collections import Counter

perguntas = [
    "Como faço a derivada de uma função?",
    "Qual é a regra da cadeia na derivada?",
    "Como resolvo um sistema linear com matrizes?",
    "O que é um autovetor de uma matriz?",
    "Como declaro uma função em Python?",
    "O que é um laço de repetição em Python?",
]

STOPWORDS = {
    "a", "o", "as", "os", "um", "uma", "de", "da", "do", "em", "na", "no",
    "que", "é", "e", "com", "qual", "como", "para", "por",
}


def tokenizar(texto):
    return re.findall(r"\w+|[^\w\s]", texto.lower(), re.UNICODE)


def so_palavras(tokens):
    return [t for t in tokens if t.isalnum()]


contagem = Counter()
for p in perguntas:
    contagem.update(so_palavras(tokenizar(p)))

print("Palavras mais frequentes (lei de Zipf em ação):")
for palavra, freq in contagem.most_common(6):
    print(f"  {palavra:12} {freq}")
```

As palavras do topo da contagem são quase todas stopwords, como a, o, de e
que. Elas dominam a frequência, mas não ajudam a distinguir um tema do outro, o que é
exatamente a motivação para removê-las.

## Removendo stopwords

Com a lista de stopwords em mãos, filtrar é direto. Vamos comparar cada pergunta antes
e depois do filtro.

```python
def remover_stopwords(tokens):
    return [t for t in tokens if t not in STOPWORDS]


for p in perguntas[:3]:
    tokens = so_palavras(tokenizar(p))
    print("original:", tokens)
    print("filtrado:", remover_stopwords(tokens))
    print()
```

Depois do filtro, cada pergunta fica reduzida aos seus termos de conteúdo,
como derivada, regra, cadeia ou matrizes. Perguntas de temas diferentes passam a ter
pouca palavra em comum, o que ajuda muito a distinguir os assuntos. Para o projeto,
meça o vocabulário e o total de tokens antes e depois do filtro, e comente quanto o
texto encolheu.
