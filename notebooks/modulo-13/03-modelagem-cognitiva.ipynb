## O que vamos fazer aqui

Em vez de uma média grosseira, vamos estimar o domínio com knowledge tracing: uma probabilidade que
se atualiza por Bayes a cada resposta, considerando deslize (errar dominando) e chute (acertar sem
dominar). Python puro.

## Knowledge tracing

A atualização tem dois passos: o bayesiano, que revisa o domínio à luz da resposta, e o de
transição, que adiciona a chance de aprender com a tentativa.

```python
def atualizar_maestria(p_dominio, correto, p_aprender=0.2, p_deslize=0.1, p_chute=0.2):
    """Knowledge tracing: atualiza a probabilidade de domínio dada uma resposta."""
    if correto:
        num = p_dominio * (1 - p_deslize)
        den = p_dominio * (1 - p_deslize) + (1 - p_dominio) * p_chute
    else:
        num = p_dominio * p_deslize
        den = p_dominio * p_deslize + (1 - p_dominio) * (1 - p_chute)
    p_cond = num / den if den else p_dominio
    return round(p_cond + (1 - p_cond) * p_aprender, 3)


p = 0.3
historico_acerta = [p]
for _ in range(5):
    p = atualizar_maestria(p, correto=True)
    historico_acerta.append(p)
print("Acertando sempre:", historico_acerta)

p = 0.3
historico_erra = [p]
for _ in range(5):
    p = atualizar_maestria(p, correto=False)
    historico_erra.append(p)
print("Errando sempre: ", historico_erra)
```

Acertando, o domínio sobe de 0,3 para perto de 1 em poucas respostas, mas de forma gradual
(por causa do chute). Errando, fica baixo (~0,23, por causa do deslize e do aprendizado pela
tentativa). Essa estimativa prudente e interpretável é muito superior à média grosseira. Para o
projeto, rastreie o domínio de várias habilidades.
